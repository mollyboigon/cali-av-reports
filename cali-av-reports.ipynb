```python
url = "https://www.dmv.ca.gov/portal/vehicle-industry-services/autonomous-vehicles/autonomous-vehicle-collision-reports/"
```

```python
import pandas as pd
import requests
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
soup.prettify()
```

```python
txt_holder = soup.find_all(class_ = "accordion-block js-accordion-block")
txt_holder
```

```python
links_a_tags = []

for each in txt_holder:
    new_each = each.find_all("a")
    links_a_tags.append(new_each)
    
```

```python
links_a_tags
```

```python
flat_a_tags = []
for sublist in links_a_tags:
    for item in sublist:
        flat_a_tags.append(item)
    
```

```python
flat_a_tags
```

```python
links = []
for a_tag in flat_a_tags:
    links.append(a_tag.get("href"))

links
```

```python
base_url = "https://www.dmv.ca.gov"
```

```python
all_links = [base_url + link for link in links]
all_links
```
